#### Código para permitir o upload de arquivos de áudio

```python
from google.colab import files

uploaded = files.upload()

if uploaded:
    audio_referencia = next(iter(uploaded))
```

#### Código para permitir a reprodução de arquivos de áudio

```python
import IPython.display as ipd
```

```python
ipd.Audio(audio_referencia)
```

#### Carregamento do modelo

```python
from TTS.api import TTS

# Carregar o modelo XTTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2")
```

#### Sintetização do áudio

```python
# Gerar fala a partir de texto
tts.tts_to_file(
    text="Este é um exemplo de síntese de fala utilizando o modelo XTTS.",
    file_path="output.wav",
    speaker_wav=audio_referencia,
    language="pt"
)

ipd.Audio("output.wav")
```
