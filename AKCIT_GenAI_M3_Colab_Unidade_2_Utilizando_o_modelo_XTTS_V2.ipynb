#### Cdigo para permitir o upload de arquivos de udio

```python
from google.colab import files

uploaded = files.upload()

if uploaded:
    audio_referencia = next(iter(uploaded))
```

#### Cdigo para permitir a reproduo de arquivos de udio

```python
import IPython.display as ipd
```

```python
ipd.Audio(audio_referencia)
```

#### Carregamento do modelo

```python
from TTS.api import TTS

# Carregar o modelo XTTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2")
```

#### Sintetizao do udio

```python
# Gerar fala a partir de texto
tts.tts_to_file(
    text="Este  um exemplo de sntese de fala utilizando o modelo XTTS.",
    file_path="output.wav",
    speaker_wav=audio_referencia,
    language="pt"
)

ipd.Audio("output.wav")
```
